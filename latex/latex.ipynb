# $\LaTeX$ শেখা 
গণিতের শিক্ষার্থীদের জন্য $\LaTeX$ একটি অপরিহার্য টুলস। গণিত হচ্ছে একধরণের ভাষা। এই ভাষা কে সুন্দর করে লেখার উপায় হচ্ছে $\LaTeX$ ব্যবহার করে লিখা। বিশেষত গাণিতিক সমীকরণ লেখার জন্য $\LaTeX$ একটি চমৎকার উপায় । $\LaTeX$ দিয়ে আমরা বিভিন্ন ডকুমেন্ট প্রস্তুত করতে পারি । যেমন CV, প্রজেক্ট, থিসিস, গবেষণা প্রবন্ধ ইত্যাদি। তাই গণিতের সকল শিক্ষার্থীদের $\LaTeX$ শেখা উচিত। এই শেখার প্রক্রিয়াটি শুরু হতে পারে প্রথমবর্ষ থেকেই।

## কিভাবে শুরু করা যায়?

<p> কম্পিউটারে একটা $\LaTeX$ সফ্টওয়ার ইনস্টল করে নিতে হবে। যেমন <a href='http://www.xm1math.net/texmaker/'>Texmaker</a>, <a href='http://kile.sourceforge.net/'> Kile</a>  ইত্যাদি । এছাড়া অনলাইনে $\LaTeX$ ব্যবহারের জন্য <a href='https://www.sharelatex.com/'> sharelatex </a> এ একাউন্ট ওপেন করতে পারেন। </p>

<p> $\LaTeX$ ব্যবহার করে আপনি কি প্রস্তুত করে চান? প্রথমে আপনার CV প্রস্তুত করতে পারেন । এর জন্য আপনি একটি CV template সংগ্রহ করতে পারেন। ইন্টারনেটে হাজার হাজার template রয়েছে। সেখান থেকে আপনার পছন্দের একটি খুঁজে বের করুন । এছাড়া আপনি যদি sharelatex ব্যবহার করেন সেখানেও রয়েছে অসংখ্য template. <a href='https://www.sharelatex.com/'> sharelatex </a> এ আপনার পছন্দের CV template থেকে তৈরি করে ফেলুন আপনার CV. CV প্রস্তুতের জন্য যে .tex ফাইল আপনি ব্যবহার করেছেন সেটি আপনার কম্পিউটারে সেভ করে রাখুন । এখন আপনার কম্পিউটারের ইনস্টল করা $\LaTeX$ সফটওয়ার ব্যবহার করে আপনার CV এডিট করা শিখুন । হ্যাঁ, আপনিও এখন একজন $\LaTeX$ ব্যবহারকারি!!! </p>

<p> LaTeX ব্যবহারের গুরুত্ব পূর্ণ দিক হচ্ছে ডকুমেন্ট প্রস্তুত করা । যেখানে ছবি, টেবিল ও গাণিতিক সমীকরণ থাকবে। এর জন্য আবারো <a href='https://www.sharelatex.com/'> sharelatex </a> ব্যবহার করতে পারেন। <a href='https://www.sharelatex.com/'> sharelatex </a> এ একটি নতুন প্রজেক্ট তৈরি করুন । যেমন প্রজেক্টটির নাম হতে পারে Quadratic Equations. এর মধ্যে আপনি নিচের লেখাটি লিখতে চান: </p>

<p> When $a \ne 0$, quadratic quation $ax^2 + bx + c = 0$ has two solutions and the roots are
$x = {-b \pm \sqrt{b^2-4ac} \over 2a}.$ </p>
<p>
$ \begin{array}{rcl}
y & = & a x^{2}+bx+c\\
  & = & a\left(x^{2}+2\times\dfrac{b}{2a}x+\dfrac{c}{a}\right)\\
  & = & a\left(x+\dfrac{b}{2a}\right)^2 - \dfrac{b^2}{4a} + c
\end{array} $ <br>
Vertex of the parabola, $y = f(x)$ is $\left(-\dfrac{b}{2a}, c - \dfrac{b^2}{4a}\right)$.
</p>

<p> এখন আপনি $\LaTeX$ ব্যবহার করে প্রজেক্ট, থিসিস বা জার্নাল পেপার লিখার জন্য প্রস্তুত হয়ে গেলেন। <a href='https://www.sharelatex.com/'> sharelatex </a> এ অসংখ্য থিসিস টেমপ্লেট রয়েছে । সেখান থেকে আপনার পছন্দের template টি বাছাই করে তৈরি করে ফেলুন আপনার থিসিস। হতে পারে সেটি এমআইটি, হারবার্ড, প্রিন্সটন কিংবা অক্সফোর্ডের থিসিস টেমপ্লেট। আর জার্নালে পেপার সাবমিট এর জন্য প্রত্যেক জার্নালের রয়েছে নিজস্ব template. সেই template এ লিখে ফেলুন আপনার প্রথম $\LaTeX$ এ লেখা পেপার। </p>
