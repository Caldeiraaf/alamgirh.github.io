# Plotting

The most important function in matplotlib is plot, which allows you to plot 2D data. 

## Power functions plots
### $y = x^n$, $\quad n \in \mathbb{Z}$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 100)

plt.plot(x, x, label='$y=x$',color='red')
plt.plot(x, x**3, label='$y=x^3$',color='green')
plt.plot(x, x**5, label='$y=x^5$',color='blue')
plt.axis('equal')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.title("Power Functions Plots ($y = x^n$, where $n = 1, 3, 5, ... $ are odd powers)\n Domain: $\mathbb{R}$ $\quad$ and $\quad$ Range: $\mathbb{R}$")

plt.axis([-3, 3, -3, 3])
plt.legend()
plt.annotate('(0, 0)', xy=(0, 0), xytext=(.1, -.3))
plt.annotate('(1, 1)', xy=(1, 1), xytext=(1.1, .9))
plt.annotate('(-1, -1)', xy=(-1, -1), xytext=(-.9, -1.1))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 100)

plt.plot(x, x**2, label='$y=x^2$',color='red')
plt.plot(x, x**4, label='$y=x^4$',color='green')
plt.plot(x, x**6, label='$y=x^6$',color='blue')
plt.axis('equal')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.title("Power Functions Plots ($y = x^n$ where $n = 2, 4, 6, ...$ are even powers) \n Domain: $\mathbb{R}$ $\quad$ and $\quad$ Range: $[0 , \infty)$")

plt.axis([-3, 3, -3, 3])
plt.annotate('(0, 0)', xy=(0, 0), xytext=(.1, -.3))
plt.annotate('(1, 1)', xy=(1, 1), xytext=(1.1, .9))
plt.annotate('(-1, 1)', xy=(-1, 1), xytext=(-2, 0.9))
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 100)

plt.plot(x, 1/x, label='$y=x^{-1}$',color='red')
plt.plot(x, 1/x**3, label='$y=x^{-3}$',color='green')
plt.plot(x, 1/x**5, label='$y=x^{-5}$',color='blue')
plt.axis('equal')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.title("Power Functions Plots ($y = x^n$, where $n = -1, -3, -5, ... $ are negative odd powers) \n Domain: $\mathbb{R}-\{0\}$ $\quad$ and $\quad$ Range: $\mathbb{R}-\{0\}$ \n Horizontal asymptote: $y =0$ ($x$ -axis) \n Vertical asymptote: $x = 0$ ($y$ - axis)")
plt.annotate('(0, 0)', xy=(0, 0), xytext=(.1, -.3))
plt.annotate('(1, 1)', xy=(1, 1), xytext=(1.1, .9))
plt.annotate('(-1, -1)', xy=(-1, -1), xytext=(-.9, -1.1))
plt.axis([-3, 3, -3, 3])
plt.legend()
plt.show()
```
